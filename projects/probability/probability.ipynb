# Обнаружение дефектов продуктов с использованием Вероятности

Вы ответственный за мониторинг кол-ва дефектных изделий на заводе. Вам сказали, что кол-во дефектов в определенный день следует распределению Пуассона с параметром скорости (лямбда), равным 7. Вы новенький, поэтому хотите почувствовать, что значит следовать распределению Пуассона, вы помните что распределение Пуассона является особенным, поскольку параметр скорости представляет собой ождаемое значение распределения, поэтому в этом случае ожидаемое значение распределения Пуассона составляет 7 дефектов в день.

Вы изучите некоторые атрибуты распределения Пуассона, чтобы получить интуитивное представление о том, сколько дефектных объектов вы можете ожидать за определенный промежуток времени. Вы также попрактикуетесь и примените то, что вы знаете о распределении Пуассона, на практическом наборе данных, который вы сами смоделируете.

```python
# Импортируем библиотеки
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

# создадим функцию для красивого отображения гистограммы
def histogram_function(rand_vars):
    plt.hist(rand_vars, bins = np.arange(len(set(rand_vars)))-0.5, edgecolor = 'black')
    plt.xticks(list(range(rand_vars.max())))
    plt.show()
    plt.close()
```

```python
# Создайте переменную lam которая представляет значение параметра распределения (lambda)
lam = 7
# Рассчитайте и выведите на экран вероятность наблюдения точно lam объектов с дефектом в заданный день
print(stats.poisson.pmf(lam, lam))
# Рассчитайте вероятность в выбранный день в который может быть 4 и менее дефектов - это хороший день
print(stats.poisson.cdf(4, lam))
# Рассчитайте вероятность в выбранный день в который может быть 9 и более дефектов - это плохой день
print(1 - stats.poisson.cdf(9, lam))
```

```python
# Создайте переменную year_defects который содежит в себе 365 случайных значений по распределению Пуассона
year_defects = stats.poisson.rvs(lam, size = 365)
# Давайте взглянем на наш набор данных
print(year_defects[:20])
```

```python
# учитывая лямбду распредления Пуассона получается что в среднем мы ожидаем в день lam дефектов в таком случае сколько всего деффектов будет за год
print(7 * 365)
# теперь взгляем сколько в общей сумме дефектов мы можем получить по распределению Пуассона
print(year_defects.sum())
# выведите на экран среднее кол-во дефектов из нашего набора
print(year_defects.mean())
# какое максимальное кол-во деффектов возможно за день
print(year_defects.max())
# да это очень редкий случай давайте рассчитаем его вероятность или более
print(1 - stats.poisson.cdf(16, lam))
# Как можно заметить по результату вероятность очень маленькая
```

```python
# Давайте рассчитаем кол-во дефектов с персентилем в 90 распределения Пуассона(7)
# Для этого используется специальный метод ppf(percentile, lam)
print(stats.poisson.ppf(.90, lam))
# При использовании данного метода мы получаем значение
# Это значение обозначает то что 90% дней у нас будет меньше кол-во дефектов чем это значение
```

```python
# А теперь давайте посмотрим, какая часть нашего смоделированного набора данных больше или равна числу, которое мы получили в предыдущем блоке
# По определению процентиля мы ожидаем что в этом диапазоне будет находится 10% дней
# Подсчитайте кол-во значений в наборе данных, которые больше или равны значению 90 процентиля
# Разделите это число на длину набора данных
print(sum(year_defects >= stats.poisson.ppf(.9, lam))/len(year_defects))
# используется функция суммы значений так как внутри данной функции указываются булевые значения где true = 1 а false = 0
```
