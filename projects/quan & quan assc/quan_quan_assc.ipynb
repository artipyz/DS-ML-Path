# Задача

- Проанализировать данные в наборе данных penguins.csv
- Построить точечный график для параметров **flipper_length_mm** и **body_mass_g**
- Проанализировать график и понять есть ли зависимость и какая она
- Расчитать значения ковариации и корреляции
- Проанализировать соответствуют ли полученные значения тому графику который был сформирован выше

```python
# Импортируем библиотеки
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import pearsonr

# Выставим точность данных полученных от np
np.set_printoptions(suppress=True, precision=1)

# Загрузим данные в переменную penguins
penguins = pd.read_csv('penguins.csv')

# Выведем первые несколько строк набора данных
print(penguins.head())
```

```python
# Создаем точечный график для двух переменных
plt.scatter(x=penguins.flipper_length_mm, y=penguins.body_mass_g)
plt.show()
plt.close()
```

Судя по полученому графику можно скачать что прослеживается линейная зависимость между указанными переменными

```python
# рассчитаем числовые показатели зависимости корреляцию и ковариацию
# для расчета ковариации будем использовать метод cov() библиотеки numpy
cov_flip_body = np.cov(penguins.flipper_length_mm, penguins.body_mass_g)
print(cov_flip_body)
# Показатель ковариации равен 9852.5
# для расчета корреляции будем использовать метод pearsonr библиотеки scipy.stats
cor_flip_body, p_flip_body = pearsonr(penguins.flipper_length_mm, penguins.body_mass_g)
print(cor_flip_body)
# Показатель корреляции довольно высок 0.8729788985653614
# Что подтверждает наличие линейной зависимости между двумя числовыми переменными
```
