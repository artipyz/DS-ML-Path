# Расчет вероятности

Так как мы знаем распределение средних значений выборок, мы можем его использовать для выявление вероятности наблюдения в определенном диапазоне средних значений выборок, которая дает некоторую информацию о всей популяции.

Давайте вернется к нашему набору данных о рыбах. Давайте представим что мы перевозим рыбу и хотим быть уверены что ящик в котором мы будем перевозить рыбу будет достаточно крепок чтобы выдержать ее вес и не сломаться.

- Так как мы уже знаем что средний вес рыбы составляет 60 кг со стандартным отклонение в 40 кг
- У нас есть ящик который выдерживает вес до 750 кг, и мы хотим перевозить в одном ящике минимум 10 рыб
- Мы хотим расчитать вероятность что средний вес всех 10 рыб будет меньше 750 кг

Используя CLT (Central Limit Theorem), мы находим средний вес 10 случайных рыб из всей популяции который составляется приблизительно 60 кг при нормальном распределении, и стандартная ошибка равна = 40 / 10^.5

Таким образом мы можем использовать вероятность распределения для расчета вероятности что вес 10-ти случайных рыб из всей популяции будет меньше 750 кг.

```python
# импортируем библиотеки
import scipy.stats as stats
# определяем переменные и их значения
x = 75 # значение веса одной рыбы
mean = 60 # среднее значение веса по популяции
std_dev = 40 # стандартное отклоенение веса по популяции
samp_size = 10 # размер выборки
standart_error = std_dev / (samp_size ** .5) # стандартная ошибка
# таким образом мы можем рассчитать вероятность используя cdf (cumulative disctribution function)
print(stats.norm.cdf(x, mean, standart_error))
# таким образом получаем вероятность 88.2% что средний вес одной рыбы составит x = 75 кг
```
