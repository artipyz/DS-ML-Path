# Случайная выборка в Python

Сгенерируем случайную выборку из популяции используя Python и отобразим график.

В этом нам помогут библиотеки:
- numpy
- pandas
- seaborn
- matplotlib

```python
# импортируем нужные библиотки
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# загрузим информацию о популяции в переменную
population = pd.read_csv('salmon_population.csv')
# изменим тип данных переменной и присвоим значение только веса
population = np.array(population.Salmon_Weight)
# найдем среднее значение популяции
pop_mean = round(np.mean(population), 3)

# Отобразим распределение графически
# формируем гистограмму
sns.histplot(population, stat='density')
# добавляем на гистограмму линию среднего значения
plt.axvline(pop_mean, color='r', linestyle='dashed')
# указываем наименование графика
plt.title(f'Population Mean: {pop_mean}')
# указываем наименование осей
plt.xlabel('Weight (lbs)')
plt.show()
plt.clf()
```

## Выборка из популяции

```python
# создаем переменную которая будет обозначать размер выборки
samp_size = 10
# создаем переменную в которой будут храниться значения выборки
# используя специальный метод библиотеки numpy random и его метод choice формируем выборку
sample = np.random.choice(np.array(population), samp_size, replace=False)
# переменная replace используется для указания того что значение котороые было взято из популяции должно быть убраны при последующих выборах данных в рамках одной выборки
# находим среднее значение 
sample_mean = round(sample.mean(), 3)

# Отображаем данные графически
sns.histplot(sample, stat='density')
plt.axvline(sample_mean,color='r',linestyle='dashed')
plt.title(F"Sample Mean: {sample_mean}")
plt.xlabel("Weight (lbs)")
plt.show()
plt.clf()
```
