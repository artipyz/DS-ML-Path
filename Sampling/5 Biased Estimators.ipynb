# Смещенные оценки
(Biased Estimators)

На основании теоремы центрального предела, среднее значение средних значений выборок равно среднему значению популяции. Это важно но не все что можно сказать о распределении выборок.

Помните, вы можете сделать распределение выборок для любых простых параметров статистики, такие как:
- среднее
- медиана
- максимальное / минимальное
- отклонение

По той простой причине что среднее значение всех средних значений выборок из популяции равно среднему значению популяции, ее называют ***несмещенной оценкой (unbiased estimator)***

В статистике любое значение называют несмещенным только в том случае если оно равно значению статистика всей популяции.

Максимальное значение это один из примеров смещенной оценки, это значит что среднее значение максимальных значений распределения выборки не централизовано в максимальном значении популяции
