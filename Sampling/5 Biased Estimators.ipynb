# Смещенные оценки
(Biased Estimators)

На основании теоремы центрального предела, среднее значение средних значений выборок равно среднему значению популяции. Это важно но не все что можно сказать о распределении выборок.

Помните, вы можете сделать распределение выборок для любых простых параметров статистики, такие как:
- среднее
- медиана
- максимальное / минимальное
- отклонение

По той простой причине что среднее значение всех средних значений выборок из популяции равно среднему значению популяции, ее называют ***несмещенной оценкой (unbiased estimator)***

В статистике любое значение называют несмещенным только в том случае если оно равно значению статистика всей популяции.

Максимальное значение это один из примеров смещенной оценки, это значит что среднее значение максимальных значений распределения выборки не равно в максимальном значению популяции

```python
# импортируем библиотеки
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Создаем переменную для названия графика
app_stat_text = 'Maximum'
# создаем функцию которая может принимать различные методы для нахождения статистики
# вы можете изменить метод нахождения максимального значения на другой: медиана, среднее, минимальное, отклонение (variance)
def app_statistics(x):
    return np.max(x)

# создаем переменные среднего значения и стандартного отклонения
mean, std_dev = 50, 15
# создаем переменную с данными
population = np.random.normal(mean, std_dev, size = 1000)

# находим максимальное значение
pop_statistics = round(app_statistics(population), 2)

# формируем график
sns.histplot(population, stat = 'density')
# добавляем на график линию которая будет показывать максимальное значение популяции
plt.axvline(pop_statistics, color='r', linestyle='dashed')
# даем название графика
plt.title(f'Population {app_stat_text}: {pop_statistics}')
# убираем название осей
plt.xlabel('')
plt.show()
plt.clf()
```

```python
# формируем график для распредления выборок
sample_stats = []
# указываем размер выборки
samp_size = 5
# с помощью цикла заполняем переменную sample_stats
for i in range(500):
    samp = np.random.choice(population, size=samp_size, replace=False)
    # переменную replace указываем False для того чтобы в одну и туже выборку не попали одни и теже значения
    sample_stats.append(app_statistics(samp))

# формируем график
sns.histplot(sample_stats, stat='density')
plt.title(f'Sampling Dist of the {app_stat_text}\nMean: {round(np.mean(sample_stats), 2)}')
# создаем линию обозначающую среднюю линию выборок
plt.axvline(np.mean(sample_stats), color='r', linestyle='dashed')
plt.xlabel(f'Sample {app_stat_text}')
plt.show()
plt.clf()
```
