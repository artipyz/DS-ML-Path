# Ожидаемое значение Биноминального Распределения

Другие типы распределения имеют ожидаемое значения или отклонение основанными на передаваемых параметрах, также как и распределение Пуассона.

Вспомним что Биноминальное распределение имеет параметры N (который представляет собой число событий) и P (который представляет собой вероятность успешного события)

Рассмотри на примере подброса монетки.

Мы подбрасываем монетку 10 раз и считаем кол-во Орлов. Как много Орлов мы ожидаем увидеть ? Возможно вы подумаете что около 5, и вы будете правы. Что мы сделали так это рассчитали ожидаемое значение без фактического проведения экспериментов. Мы взяли кол-во экспериментов (10) и перемножили это значение на вероятность (.5)

Прошу обратить внимание что это нормально когда есть дробная часть в значении ожидаемого значения / отклонения

## Отклонение биноминального распределения
Отклонение биноминального распределения это значение на сколько ожидаемое значение успеха события может отклонятся, другими словами измерение диапазона вероятностни успеха среднего/ожидаемого значения.

Отклонение биноминального распределения расчитывается также как и ожидаемое значение с использованием значений параметров N (кол-во экспериментов) и P (вероятность ождаемого события).

Также рассмотрим рассчет на примере подброса монетки. Давайте также подбросим монетку 10 раз, для того чтобы понять как рассчитать отклонение. Каждый подброс монетки имеет вероятность .5 на каждую из свох сторон. Отклонение каждого отдельного подброса монеты рассчитывается как произведения ожидаемого значения и вероятности не успешного события

Таким образом отклонение ожидаемого значения подброса монетки при 10 экспериментах является

`10 * 0.5 * (1 - 0.5)`

Где:
- 10 * .5 - это ожидаемое значение
- 1 - .5 - это вероятность того что событие на произойдет

Рассмотрим тот же рассчет на примере выбора правильного варианта ответов из 20 вопросов в каждом вопросе 4 варианта ответа. Если учесть что в каждом вопросе может быть только 1 правильный ответ таким образом если мы ответи на все вопросы только один вариантом ответа в таком случае ожидаемое значение будет равно `20 * .25` а для того чтобы рассчитать отклонение `20 * .25 * (1-.25)=3.75`

Другими словами мы ожидаем увидеть 5 правильных ответов но с учетом отклонение данного значения, правильных ответов может быть от 2 до 8

# Свойства Ожидаемого значения и его отклонения

## Свойства ожидаемого значения

1. Ожидаемое значение двух независимых случайных переменных рассчитывается как сумма ожидаемых значений этих переменных

Пример. Если мы хотим посчитать общее кол-во орлов между подрасыванием четвертака 10 раз и пенни 6 раз, ожидаемое значение будет складываться из 5 орлов (четвертака) и 3 орлов пенни таким образом получаем ожидаемое значение равное 8.

2. Перемножение случайного значения на неизменное значение (constant), изменит ожидаемое значения в такое же кол-во раз как и значение.

Пример. Ожидаемое значение Орлов при подбросе монеты 10 раз равно 5. Если мы хотим рассчитать ождаемое значение для события которое будет выполненно 4 раза (10 подробросов 1 событие таким образом общее кол-во подбросов 40), ожидаемое значение также возрастет в 4 раза и получаем 20.

3. Добавление неизменного значения (constant) к распределению изменяет ожидаемое значение на величину этой константы

Пример. Предположим что студенты прошли тест и среднее значение результата этого теста составило 78 баллов из 100. Таким образом если мы добавим 2 балла к каждой оценке в распределении, то средняя оценка также возрастет на два получается станет равно 80

## Свойства отклонения

1. Увеличение значений в распределении на константу ни как не повлият на значение отклонения

Объяснение. Это происходит потому что отклонение самой константы на значение которой будут увеличены значения распределения равно 0.

2. Масштабирование значений случайных величин на значение константы масштабирует отклонение на величину константы в квадрате
3. Отклонение суммы двух случайных переменных явялется суммой отклонений каждой переменной
