# Дискретные и непрерывные случайные величины
(Discrete and Continuous Random Variables)
## Дискретные случайные величины
(Discrete Random Variables)

Случайные величины с подсчитанным числом возможных значений называют дискретными

Например:
- Бросок игральной кости: Так как мы можем посчитать кол-во возможных результатов есть граней 6 то и значений 6

Дискретные случайные величины также часто применяются когда иследуются счетные события

Например:
- как много людей зашло в магазин в случайный день
- данный случай подойдет потому что мы не сможем считать людей по половинам

## Непрерывные случайные величины
(Continuous Random Variables)

Когда мы не можем посчитать кол-во возможных результатов такие величины называют непрерывными

Обычно это величины которые можно измерить и являются дробями (имеют дробную часть):
- длинна
- высота
- и т.д.
