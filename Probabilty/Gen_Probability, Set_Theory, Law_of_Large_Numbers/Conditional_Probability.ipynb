# Условная вероятность (Conditional Probability)
Если мы хотим вычислить вероятность возникновения двух зависимых событий, нам следует для начала определить условную вероятность.

Давайте вспомним какие события являются зависимыми
- Зависимыми событиями называют такие события результат возникновения которых зависит от результатов предыдущих событий

Например
- Возьмем мешок с шариками в ктором находятся 3 красных и 2 синих шарика
- Какова вероятность вытащит красный шарик вторым если
    - Мы вытащим первым синий шарик и не будем возвращать его в мешок
    - Мы вытащим первым красный шарик и не будем возвращать его в мешок

Для таких событий используется понятие условной вероятности которая зависит от результатов предыдущих событий

Для обозначения условной вероятности используют вертикальную черту
- P (Red Second | Blue First)

Другими словами мы записали - *какова вероятность того что вторым шариком будет красный если первым мы вытащили синий?*

## Итог

В итоге результатом первых условий
- Если первым вытащим синий шарик и не будем возвращать его в мешок

```P (Red Second | Blue First) = 3/4```

Результатов вторых условий
- Если первым вытащим красный шарик и не будем возвращать его в мешок

`P (Red Second | Red First) = 2/4`

## А что если

А что если мы будем вытаскивать шарики и класть их обратно в мешок в таком случае события станут не зависимыми и вероятность будет рассчитываться как

`P (Red Second | Blue First) = P(Red)`
