# Теорема Байеса

![Bayes Theorem Tree Diagram](bayes%20theorem%20tree%20diagram.png)

Представьте что вы пациент который недавно получил положительный тест на ангину
- Вы хотите узнать а какова вероятность что у вас есть ангина
$$
P(ST|+)
$$
Для того чтобы рассчитать вероятность этого события, мы будем использовать теорему Байеса, которая представляется формулой
$$
P(B|A)=\frac{P(A|B)*P(B)}{P(A)}
$$
Используя данную теорему теперь мы можем узанть вероятность
$$
P(ST|+) = \frac{P(+|ST)*P(ST)}{P(+)}
$$
Мы также знаем из диаграммы выше что:
$$
P(+|ST)=0.85
P(ST)=0.20
$$
Но что мы знаем о значении P(+)?

Давайте рассуждать:
- Мы можем получить 4 различных результата
    - Есть ангина и тест положительный
    - Есть ангина и тест отрицательный
    - Нет ангины и тест положительный
    - Нет ангины и тест отрицательный

В данном случае нас интересуют ситуации только с положительным тестом

Получается что:
$$
P(+)=P(ST and +) + P(NO ST and +)=0.17 + 0.016=0.186
$$
Теперь у нас есть все данные для расчета вероятности события P(ST|+) которого звучит как: Какова вероятность того что я болею ангиной при получении положительного теста на ангину
$$
P(ST|+) = \frac{0.85 * 0.20}{0.186} = 0.914
$$

Таким образом получаем вероятность ***91,4%*** что вы болеете ангиной при получении положительного результата теста

## Задача

Самостоятельно расчитать вероятность того что вы болеете ангиной при получении отрицательного теста

Надо расчитать вероятность ST при том что результат предыдущего события был отрицательный тест
$$
P(ST|-)
$$
Формула Байеса будет выглядеть следующим образом
$$
P(ST|-)=\frac{P(-|ST)*P(ST)}{P(-)}
$$
Что мы знаем:
- P(-|ST) = 0.15
- P(ST) = 0.20
- P(-) = P(ST and -) + P(NO ST and -) = 0.03 + 0.784 = 0.814

Таким образом получаем
$$
P(ST|-)=\frac{0.15 * 0.20}{0.814}=0.037
$$
Вероятность что вы более ангиной при отрицательном тесте на ангину равно ***3.7%***

# ИТОГ

Такис образом Теория Байса позволяет нам найти обратную вероятность от полученных результатов
