# Правило перемножения событий (Multiplication Rules)

Ранее мы рассматривали правило сложения вероятностей, которое описывает вероятность возникновения одного или другого события по отдельности.

Что если нам потребуется рассчитать вероятность возникновения двух событий одновременно?

Или другими словами рассчитать вероятность пересечения значений событий (*Intersection*)

Основной формулой расчета такой веростности выглядит следующим образом:

`P (A and B) = P(A) * P(B|A)`

**Другими словами умножение вероятности события A и зависимого события B при том что событие A произошло**

## Зависимые события (Dependent Events)

Рассмотрим формулу расчета возникновения двух зависимых событий одновременно на примере мешка с шарами

Условия:
- У нас есть мешок с 5 шарами
- 2 шарика синих
- 3 шарика красных
- Мы берем два шарика без возврата их в мешок

Задача:
- Рассчитать вероятность того что мы вытащик синий шарик в первый и второй раз

Вероятность такого события рассчитывается по формуле

```
P (Blue First and Blue Second) = P(Blue First) * P(Blue Second | Blue First)
P (Blue First and Blue Second) = 2/5 * 1/4
P (Blue First and Blue Second) = 1/10
```

Таким образом вероятность таких событий будет равно *10%*

Это один из возможных результатов когда мы берем два шарика из мешка.

## Независимые события (Independent Events)

Для двух независимых событий, правило умножения становится более упрощенным. Формула расчета возникновения двух независимых событий одновременно выглядит следующим образом

`P(A and B) = P(A) * P(B)`

Так происходит потому как событие не зависимо а при не зависимых событих формула условной вероятности приводится к простому виду

`P(B | A) = P(B)`

Рассмотрим это на примере подброса монетки

Если мы подбросим монетку два раза какова вероятность что она дважды упадет Орлом давайте рассчитаем
- При первом броске вероятность 0.5
- При втором броске вероятность также 0.5

Потому что события являются независимыми. Таким образом

`P(A and B) = P(A) * P(B) = 0.5 * 0.5 = 0.25`
